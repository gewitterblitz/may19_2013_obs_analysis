### Trying to match the lma grid with KTLX grid (generated from RadX) plots

I realized that the lma gridded products generated originally in the `grid_files20002300_per_radar_grid_500x500x500m_KTLX_per_PyART` folder </br>
used x_bnd, y_bnd, and z_bnd limits which do not overlap with the bounding limits of RadX generated grid files.

Therefore, it may be a good idea to generate flash products again with the new grid limits </br>
> `x_bnd=(-74.75e3, 45.25e3), y_bnd=(0.25e3, 120.25e3), z_bnd=(0.25e3, 20.25e3)`

> The grid centers for flash products will therefore match with the RadX grid centers
